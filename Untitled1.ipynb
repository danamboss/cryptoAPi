```python
def api_runner():
  global df
  url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
  parameters = {
    'start':'1',
    'limit':'15',
    'convert':'USD'
  }
  headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': 'eebd5c06-c39b-49e9-8c59-92f07cff57b7',
  }

  session = Session()
  session.headers.update(headers)

  try:
    response = session.get(url, params=parameters)
    data = json.loads(response.text)
    print(data)
  except (ConnectionError, Timeout, TooManyRedirects) as e:
    print(e)

  df2 = pd.json_normalize(data['data'])
  df2['timestamp'] = pd.to_datetime('now')
  df = df.append(df2)

  #if not os.path.isfile(r'C:\Users\ambos\Downloads\script\Api.csv'):
     # df.to_csv(r'C:\Users\ambos\Downloads\script\Api.csv', header='column_names')
 # else:
      #df.to_csv(r'C:\Users\ambos\Downloads\script\Api.csv', mode='a', header=False)"""
```
